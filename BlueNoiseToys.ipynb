```python
# Take some of the blue noise textures and tile them 8x8 to see how it looks. Blue noise textures should be resistant to banding with tiling.

tiled_width = width * 8
tiled_height = height * 8
blue_noise_texture_2x2_tiled = np.zeros((tiled_height, tiled_width), dtype = np.uint8)
blue_noise_texture_4x4_tiled = np.zeros((tiled_height, tiled_width), dtype = np.uint8)

for i in range(0, tiled_width, width):
	for j in range(0, tiled_height, height):
		blue_noise_texture_2x2_tiled[j : j + height, i : i + width] = blue_noise_texture_2x2
		blue_noise_texture_4x4_tiled[j : j + height, i : i + width] = blue_noise_texture_4x4


# Create the visualization
plt.figure(figsize=(10, 10))
plt.imshow(blue_noise_texture_2x2_tiled)
plt.title('2x2 blue noise tiled')
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.tight_layout()
plt.show()
```

```python
# Create the visualization
plt.figure(figsize=(10, 10))
plt.imshow(blue_noise_texture_4x4_tiled)
plt.title('4x4 blue noise tiled')
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.tight_layout()
plt.show()
```
